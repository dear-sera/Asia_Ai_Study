### DL_day1 정리

### 딥러닝 - 사람의 뇌를 구성하는 뉴런처럼 인공신경망을 구성하고, 인간과 비슷한 방식으로 학습시켜서 분류, 예측등의 머신러닝을 수행 하는 것  

#### tenserflow - 구글에서 제공하는 연구와 개발을 위한 오픈소스 딥러닝, 머신러닝 라이브러리

• 텐서(Tenser) - 다차원 배열  
텐서 scalar, vector, matrix, 3차원 이상 확장 가능한 구조    
텐서의 차원의 수 확인 - Rank, tf.rank()

• 텐서보드의 Data Flow Graph :  
노드 - 데이터를 통해 수행하는 연산 역할  
엣지 - 데이터 

#### 케라스 - 저차원의 딥러닝 라이브러리를 래핑한 고차원의 라이브러리    

• Sequential - 딥러닝의 구조를 한 층 한 층 순차적으로 층을 쌓은 신경망 모델  

• Dense - 신경망 모델에 포함된 완전 연결층  

Dense클래스에 각 층이 어떤 특성을 가질지 옵션을 설정:  
- activation - 다음 층으로 어떻게 값을 넘길지 결정하는 부분(relu, sigmoid, softmax등)  
- loss - 한 번 신경망이 실행될 때마다 오차 값을 추적하는 함수(mse, rmse등)   
- optimizer - 오차를 어떻게 줄여 나갈지 정하는 함수(경사하강법, 확률적 경사하강법 등)     
- units - 뉴런 또는 출력 노드의 개수  
- inuput_shape - 입력 데이터의 형태


에포크 (epoch) : 주어진 데이터를 신경망 모델에서 한 번 훈련하는 단위


#### 딥러닝 동작 원리

• 딥러닝을 이용해서 단순 선형 회귀 수행은 데이터를 가장 잘 설명하는 직선인 1차함수의 기울기값 절편을 찾는 것  

• 최소제곱법 - 입력 데이터가 1차원이면 기울기와 절편을 구하는 공식  
(x-x평균)(y-y평균)의 합/(x-x평균)^2의 합  

• 딥러닝에서 입력 데이터는 다차원이므로 기울기와 절편을 찾기 위해서는 임의의 선을 그리고 난 후, 이 선이 얼마나 잘 그려졌는지를 평가하여 조금씩 수정해 가는 방법을 사용합니다  

1. 가설을 하나 세운 뒤
2. 주어진 요건을 충족하는지 판단
3. 조금씩 변화를 주고, 오차가 최소가 될 때까지 과정을 계속 반복  
각 선의 오차를 계산할 수 있어야 하고, 오차가 작은 쪽으로 바꾸는 알고리즘이 필요

• 기울기와 오차와의 관계 특성은 기울기가 무한대로 커지면 와도 무한대로 커지는 상관관계는 2차함수 그래프로 표현할 수 있다  

• 미분 기울기를 이용하는 경사하강법 - 이차 함수 그래프에서 오차를 비교하여 가장 작은 방향으로 이동시키는 방법

• 미분은 '순간 변화율'을 구하는 것  
미분 계수 - 어느 순간에 어떤 변화가 일어나고 있는지를 숫자로 나타낸 것  

순간 변화율은 x의 증가량(Δx)이 0에 가까울 만큼 아주 작을 때의 순간적인 기울기  

- f(x) = a에서 a가 상수일 때 미분 값은 0
- f(x) = x일때의 미분 값은 1
- f(x) = ax에서 a가 상수이면 미분 값은 a
- f(x) = X^a에서 a가 자연수이면 미분 값은 aX^a-1

• 편미분 - 한 가지 변수만 미분하고 그 외에는 모두 상수로 취급

#### tenserflow와 keras를 이용해서 경사하강법 적용 기울기와 절편을 찾는 실습  

• 이진분류를 수행하려면 시그모이드 함수를 사용,  
 시그모이드 함수의 기울기와 절편을 경사 하강법으로 계산  
- 예측 값이 1일 때 오차는 0이고, 예측 값이 0에 가까울수록 커지는 로그 함수 그래프로 표현 -logaX
- 예측 값이 0일 때 오차는 없으면, 1에 가까워질수록 오차가 매우 커지는 로그 함수 그래프로 표현 -loga(1-X)
- 오차 = 평균 (-logaX + (1-y) -loga(1-X))


• 퍼센트론(신경망):  
입력 값과 활성화 함수를 사용해 출력 값을 다음으로 넘기는 가장 작은 신경망 단위  
- N개의 이진수가 하나의 뉴런을 통과해서 가중합 0보다 크면 활성화되는 가장 간단한 신경망 구조
- AND 게이트 , OR 게이트를 만들 수 있다.(입력데이터의 가중합 -> 활성화함수)   
- 두 개의 공간을 분리시키는 역할을 한다
- 가중합 - 입력 값(x)과 가중치(w)의 곱을 모두 더한 값에 바이어스(b)를 더한 값
- 가중합의 결과를 놓고 1 또는 0을 판단하는 활성화 함수 적용, 결과를 다음층 또는 출력으로 보냅니다  

• XOR는 직선을 그어 두 개의 공간을 분리시킬 수 없고, 데이터를 표현하는 평면 자체에 변화를 주어 해결   
- 은닉층을 만들어 공간을 왜곡하면 두 영역을 가로지르는 선이 직선으로 바뀝니다  
- 입력데이터의 가중합 -> 활성화함수 -> 은닉층 -> 가중합 -> 은닉층의 활성화함수 -> 출력 => <신경망 1 epoch>

• 신경망 내부의 가중치는 출력 -> 은닉층 -> 입력층 순서로 수정하는 방법 - 오차 역전파 방법  
가중치에서 기울기를 빼도 값의 변화가 없을   때까지 계속해서 수정 작업을 반복  

• 활성화 함수 시그모이드는 층이 늘어나면서 기울기가 중간에 0이 되어버리는 기울기 소실 문제가 발생 

• 기울기가 사라지는 문제를 해결하기 위해 활성화 함수:    
ReLU - x가 0보다 작을 때는 모든 값을 0으로 처리하고, 0보다 큰 값을 x값 그대로 사용하는 방법    
x가 0보다 크기만 하면 미분 값이 1이 됩니다    
따라서 여러 은닉층을 거치며 곱해지더라도 맨 처음 층까지 사라지지 않고 남아있을 수 있습니다    

• (오차 수정 방법) 경사 하강법 단점 : 정확하게 가중치를 찾아가지만, 한 번 업데이트 할 대마다 전체 데이터를 미분해야 하므로 계산량이 매우 많다  

• 확률적 경사 하강법 : 전체 데이터를 사용하는 것이 아니라, 랜덤하게 추출한 일부 데이터를 사용, 속도가 빠르면서도 최적 해에 근사한 값을 찾아가는 방식 (중간 결과의 진폭이 크고 불안정해 보일 수도 있습니다)  

• 오차를 수정하기 전 바로 앞 수정 값과 방향을 참고하여 같은 방향으로 일정한 비율만 수정되게 하는 방법 : 모멘텀 SGD  

• 네스테로프 모멘텀 : 변수의 업데이트가 잦으면 학습률을 적게 하여 이동 보폭을 조절하는 방법 (보폭크기개선)  

• 변수의 업데이트 횟수에 따라 학습률을 조절하는 옵션이 추가된 최적화 방법 : 아다그리드   
아다그리드는 많이 변화하지 않는 변수들을 학습률은 크게하고, 반대로 많이 변화한 변수들에 대해서는 학습률을 적게 합니다  

• 알엠에스프롭 - 아다그리드의 G(t)의 값이 무한히 커지는 것을 방지하고자 제안된 방법 

• 아담(Adam) - 모멘텀과 알엠에스프롭 방법을 합친 경사하강법

````python
from keras.models import Sequential
from keras.layers import Dense
import numpy
import tensorflow as tf

numpy.random.seed(0) # seed 값 생성
tf.random.set_seed(0)

# 데이터 로드
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",") 
X = dataset[:,0:8]
Y = dataset[:,8]

# 모델 설정
model = Sequential() 
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))  #첫 번째 층에서 두 번째 출력층으로
model.add(Dense(1, activation='sigmoid'))  #두 번째 층에서 마지막 출력층으로

# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=10) 
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1])) 
# 결과 =  Accuracy: 0.7578
````

````python
# 다중 분류 분석 tenserflow 수행 실습

df = pd.read_csv('iris.csv', names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
                 
#그래프로 확인
sns.pairplot(df, hue='species')
plt.show()
````

````python
# 데이터 특성과 label 분리
dataset = df.values
X = dataset[:, 0:4].astype(float)
Y_obj = dataset[:, 4]

# 문자열을 숫자로 반환
from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
y_encoded = tf.keras.utils.to_categorical(Y)

# 입력데이터의 은닉층으로의 활성화 함수는 relu
# 은닉층에서 출력층으로의 활성화 함수는 softmax - 총합이 1인 형태로 바꿔서 계산해 주는 함수
# 손실함수는 categorical_crossentropy, 오차 줄이는 함수 Adam, 성능지표 accuracy
# 모델 설정
model = Sequential() 
model.add(Dense(16, input_dim=4, activation='relu')) 
model.add(Dense(3, activation='softmax'))

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# 모델 실행
model.fit(X, y_encoded, epochs=50, batch_size=1)
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, y_encoded)[1])) 
# 결과 =  Accuracy: 0.9800
````

#### 과적합 피하기

• 과적합(over fitting) - 모델이 학습 데이터셋 안에서는 일정 수준 이상의 예측 정확도를 보이지만, 새로운 데이터에 적용하면 잘 맞지 않는 것

• 과적합은 층이 너무 많거나 변수가 복잡해서 발생하기도 하고 테스트셋과 학습셋이 중복될 때 생기기도 합니다

• 과적합을 방지 방법 - 학습을 하는 데이터셋과 이를 테스트할 데이터셋을 완전히 구분한 다음 학습과 동시에 테스트를 병행하며 진행

• 머신러닝의 최종 목적 - 과거의 데이터를 토대로 새로운 데이터를 예측하는 것
새로운 데이터에 사용할 모델을 만드는 것

• 학습셋만 가지고 평가할때, 층을 더하거나 에포크(epoch) 값을 높여 실행 횟수를 늘리면 정확도가 계속해서 올라갈 수 있습니다.

• 학습 데이터셋만으로 평가한 예측 성공률이 테스트셋에서도 그대로 나타나지는 않습니다.

• 학습이 깊어져서 학습셋 내부에서의 성공률은 높아져도 테스트셋에서는 효과가 없다면 과적합이 일어나고 있는 것입니다.

과적합을 줄이기 위한 데이터셋 분리 실습 



````python
# 초음파 광물 예측 분석(학습셋과 테스트셋 구분) 실습

from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy
import tensorflow as tf

seed = 0
numpy.random.seed(seed) # seed 값 설정
tf.random.set_seed(seed)

df = pd.read_csv('sonar.csv', header=None)
dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

# 입력데이터의 은닉층으로의 활성화 함수는 relu, 출력 units는 24 
# 은닉층에서 은닉층으로의 활성화 함수는 relu, 출력 units는 10
# 은닉층에서 출력층으로의 활성화 함수는 sigmoid 
# 손실함수는 mean_squared_error, 오차 줄이는 함수 Adam, 성능지표 accuracy
# 모델 설정
model = Sequential() 
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))  
model.add(Dense(1, activation='sigmoid'))

# 모델 컴파일
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
  
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=5)
  
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))

#결과 100% = 과적합
````

````python
# 학습 셋과 테스트 셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)

model = Sequential() 
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))  
model.add(Dense(1, activation='sigmoid'))  

# 모델 컴파일
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
  
# 모델 실행
model.fit(X_train, Y_train, epochs=130, batch_size=5)
  
# 결과 출력
print("\n Test Accuracy: %.4f" % (model.evaluate(X, Y)[1]))

# 모델 저장
model.save('my_model.h5')
````

````python
# 테스트를 위해 메모리 내의 모델을 삭제
del model 

from keras.models import load_model
# 모델을 새로 불러옴
model = load_model('my_model.h5') 
print("\n Test Accuracy: %.4f" % (model.evaluate(X, Y)[1]))

# 결과 = Accuracy: 0.9519
````

#### k겹 교차 검증
• k겹 교차 검증(k-fold cross validation) - k겹 교차 검증이란 데이터셋을 여러 개로 나누어 하나씩 테스트셋으로 사용하고 나머지를 모두 합해서 학습셋으로 사용하는 방법

• 데이터가 충분치 않은 경우, 데이터의 100%를 테스트셋으로 사용할 수 있습니다.

초음파 광물 예측 분석 - k겹 교차 검증

10개의 파일로 쪼개 테스트하는 10-fold cross validation을 실시하도록 n_fold의 값을 10으로 설정한 뒤 StratifiedKFold() 함수에 적용했습니다.   
그런 다음 모델을 만들고 실행하는 부분을 for 구문으로 묶어 n_fold만큼반복되게 합니다.

````python
# 초음파 광물 예측 분석 - k겹 교차 검증

from sklearn.model_selection import StratifiedKFold

n_fold = 10
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=0)

accuracy = [] # 빈 accuracy 배열

for train, test in skf.split(X, Y): # 모델의 설정, 컴파일, 실행
  model = Sequential()
  model.add(Dense(24, input_dim=60, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
  model.fit(X[train], Y[train], epochs=100, batch_size=5)
  k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])
  accuracy.append(k_accuracy)

# 결과 출력
print("\n %.f fold accuracy:" % n_fold, accuracy)

# 결과 = 10 fold accuracy: ['0.7143', '0.8571', '0.7619', '0.9524', '0.8571', '0.7143', '0.8571', '0.8571', '0.9000', '0.8500']
````

#### MNIST 손글씨 인식하기

• MNIST 데이터셋은 미국 국립표준기술원(NIST)이 고등학생과 인구조사국 직원 등이 쓴 손글씨를 이용해 만든 데이터로 구성

• 70,000개의 글자 이미지에 각각 0부터 9까지 이름표를 붙인 데이터셋

• 딥러닝을 이용 손글씨 이미지 분류 , 정확률 예측 실습

````python
# MNIST 손글씨 인식하기: 데이터 전처리

from keras.datasets import mnist
from keras.utils import np_utils
import numpy
import sys
import tensorflow as tf

seed = 0
numpy.random.seed(seed) # seed 값 설정
tf.random.set_seed(seed)

# MNIST 데이터셋 불러오기
(X_train, Y_class_train), (X_test, Y_class_test) = mnist.load_data()
print("학습셋 이미지 수 : %d 개" % (X_train.shape[0]))
print("테스트셋 이미지 수 : %d 개" % (X_test.shape[0]))

# 그래프로 확인
plt.imshow(X_train[0], cmap='Greys')
plt.show()

# 코드로 확인
for x in X_train[0]:
  for i in x:
    sys.stdout.write('%d\t' % i)
  sys.stdout.write('\n')
````

````python
# 차원 변환 과정
# Reshape() 함수를 사용하여 가로 28, 세로 28의 2차원 배열을 784개의 1차원 배열로 바꿔 주어야 합니다.
X_train = X_train.reshape(X_train.shape[0], 784)
X_train = X_train.astype('float64')
X_train = X_train / 255
X_test = X_test.reshape(X_test.shape[0], 784).astype('float64') /255

print("class : %d " % (Y_class_train[0])) # 클래스 값 확인

# 바이너리화 과정
Y_train = np_utils.to_categorical(Y_class_train, 10)
Y_test = np_utils.to_categorical(Y_class_test, 10)
print(Y_train[0])
````

````python
# MNIST 손글씨 인식하기: 딥러닝 실행

from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint,EarlyStopping
import matplotlib.pyplot as plt
import os

# 모델 프레임 설정
model = Sequential()
model.add(Dense(512, input_dim=784, activation='relu'))
model.add(Dense(10, activation='softmax'))

# 모델 실행 환경 설정
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 최적화 설정
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
  os.mkdir(MODEL_DIR)

modelpath="./model/{epoch:02d}-{val_loss:.4f}.hdf5"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10)

# 모델의 실행
history = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=30, batch_size=200, verbose=0,
callbacks=[early_stopping_callback,checkpointer])

# 테스트 정확도 출력
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))

# 테스트셋의 오차
y_vloss = history.history['val_loss']
````

````python
# 학습셋의 오차
y_loss = history.history['loss']

# 그래프로 표현
x_len = numpy.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='.', c="red", label='Testset loss')
plt.plot(x_len, y_loss, marker='.', c="blue", label='Trainset loss')

# 그래프에 그리드를 주고 레이블을 표시
plt.legend(loc='upper right')
# plt.axis([0, 20, 0, 0.35])
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')

plt.show()
````

#### 컨볼루션 신경망(Convolutional Neural Network, CNN)

• 기본 딥러닝 프레임에 이미지 인식 분야에서 강력한 성능을 보이는 신경망

• 입력된 이미지에서 다시 한번 특징을 추출하기 위해 마스크(필터, 윈도 또는 커널이라고도 함)를 도입하는 기법

• 입력층과 출력층 사이의 중간층(은폐층)에 합성곱과 풀링층을 배치한 것

• 이미지를 흐리게 만들거나 경계를 강조하는 작업을 함

• 합성곱층과 풀링층에서는 해상도를 낮추거나 샘플링하는 처리를 계속 반복함

• padding은 결과 값이 작아지는 것을 방지하기 위해서 입력값 주위로 0 값
을 넣어서 입력 값의 크기를 인위적으로 키워서, 결과값이 작아지는 것을 방지 하는 기법이다.

• 패딩은 결과 값을 작아지는 것을 막아서 특징이 유실되는 것을 막는 것 뿐 아니라, 오버피팅도 방지

• 특징을 추출하는 기능을 하는 필터(Filter)와 필터의 값을 비선형 값으로 바꾸어 주는 Activation함수로 구성

• 필터들을 통해서 Feature map이 추출되었으면, 이 Feature map에 Activation function을 적용

컨볼루션 신경망

• 합성곱층은 이미지의 특징을 추출할 때 사용

• 합성곱층의 역할은 주변의 값과 필터를 사용해 중앙에 있는 값을 변화시키는 것

• 풀링층은 합성곱층으로 얻은 특징맵 c를 축소하는 층 – max pooling, average pooling

• 전결합층은 각 층의 유닛을 결합

• 마스크를 여러 개 만들 경우 여러 개의 컨볼루션이 만들어집니다.

• 케라스에서 컨볼루션 층을 추가하는 함수는 Conv2D()입니다.

```
model.add(Conv2D(32, kernel_size=(3, 3), input_shape=(28, 28, 1), activation=‘relu’))
```

• 첫 번째 인자 : 마스크 적용 개수  
• kernel_size : 마스크(커널)의 크기. kernel_size=(행, 열) 형식  
• input_shape : 맨 처음 층의 입력 뉴런 수  
• activation : 활성화 함수 정의

• 풀링 (Sub sampling or Pooling) : 추출된 Activation map을 인위로 줄이는 작업
컨볼루셔널 계층을 통해서 어느정도 특징이 추출 되었으면, 이 모든 특징을 가지고 판단을 할 필요가 없다.

• max pooling, average pooling, L2-norm pooling 등 (max pooling 이라는 기법이 많이 사용된다.)

• Max pooling (맥스 풀링) : Activation map을 MxN의 크기로 잘라낸 후, 그 안에서 가장 큰 값을 뽑아내는 방법

• sampling 을 통해서 얻을 수 있는 장점 :  
전체 데이타의 사이즈가 줄어들기 때문에 연산에 들어가는 컴퓨팅 리소스가 적어지고
데이타의 크기를 줄이면서 소실이 발생하기 때문에, 오버피팅을 방지할 수 있다.

컨볼루션 신경망(Convolutional Neural Network, CNN) – 과적합 피하기

• 딥러닝 학습을 실행할 때 가장 중요한 것은 과적합을 얼마나 효과적으로 피해가는지에 달려 있습니다.

• 드롭아웃(drop out) 기법 - 은닉층에 배치된 노드 중 일부를 임의로 꺼주는 것입니다.

• 랜덤하게 노드를 끔으로써 학습 데이터에 지나치게 치우쳐서 학습되는 과적합을 방지할 수 있습니다.

```
model.add(Dropout(0.25)) #25%의 노드를 끔
```


````python
from keras.datasets import mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D,MaxPooling2D
from keras.callbacks import ModelCheckpoint,EarlyStopping
import matplotlib.pyplot as plt
import numpy
import os
import tensorflow as tf

# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(seed)

# 데이터 불러오기
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 28, 28,1).astype('float32') / 255
X_test = X_test.reshape(X_test.shape[0], 28, 28,1).astype('float32') / 255
Y_train = np_utils.to_categorical(Y_train)
Y_test = np_utils.to_categorical(Y_test)
````

````python
# 컨볼루션 신경망 설정
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), input_shape=(28, 28, 1),activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 최적화 설정
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
  os.mkdir(MODEL_DIR)

modelpath="./model/{epoch:02d}-{val_loss:.4f}.hdf5"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=10)
````

````python
# 모델의 실행
history = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=30, batch_size=200, verbose=0,
callbacks=[early_stopping_callback,checkpointer])

# 테스트 정확도 출력
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))

# 테스트셋의 오차
y_vloss = history.history['val_loss']

# 학습셋의 오차
y_loss = history.history['loss']

# 그래프로 표현
x_len = numpy.arange(len(y_loss))
plt.plot(x_len, y_vloss, marker='.', c="red", label='Testset_loss')
plt.plot(x_len, y_loss, marker='.', c="blue", label='Trainset_loss')

# 그래프에 그리드를 주고 레이블을 표시
plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
````
