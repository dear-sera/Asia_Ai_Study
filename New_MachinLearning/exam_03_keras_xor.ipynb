## keras_xor

이진분류기 (맞다, 아니다)

XOR = 같지 않으면 1 같으면 0

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
#학습, 타켓 데이터 설정 

training_data = np.array([[0,0], [0, 1], [1, 0], [1, 1]], 'float32')  #'float32'타입으로 전환
target_data = np.array([[0], [1], [1], [0]], 'float32')  # xor연산에 대해 윗줄과 비교한 정답값
```

```python
#모델 생성

model = Sequential()
model.add(Dense(32, input_dim=2, activation='relu'))  #Dense(32= 퍼셉트론(뇌세포가 32개), input_dim=2= input값이 2개(XOR은 이진분류기여서 2개)
model.add(Dense(1, activation='sigmoid'))   #활성화 함수 sigmoid = 이진분류기에 대해 미분을 해주는 함수(맨끝층에만 사용)
model.compile(loss='mse', optimizer='adam', metrics=['binary_accuracy'])  #옵티마이저는 주로 아담을 쓴다

print(model.summary())   #summary = 요약으로 보여준다

#파라미터 값 = (노드의 수) * (input의 차원) + (노드의 수)
#dense (Dense)                (None, 32)                96   => 첫 레이어에서 96은 64개의 input_dim으로 온 것들(노드의 수 * input차원) + 32개의 노드의 수
#dense_1 (Dense)              (None, 1)                 33   => 두 번째 레이어에서 1개의 input의 차원과 32개의 노드의 수 
```

input_dim 은 2개의 input이 신경세포로 뻗어나감

```python
fit_hist = model.fit(training_data, target_data, epochs=500, verbose=1)  #verbose=1
```

```python
plt.plot(fit_hist.history['loss'])  #오차가 얼마나 줄어드는지 확인
plt.show()
```

```python
inp = list(map(int, input().split()))
qwe = np.array(inp)
print('입력 값 : ', qwe)
qwe = qwe.reshape(1, 2)
print('reshape : ', qwe)
print('결과 값 : ', model.predict(qwe)[0][0].round())

#결과 값 :  0.0 -> 맨위 입력값 1,1에 대한 타겟이 0인 것에 대한 예측
```
