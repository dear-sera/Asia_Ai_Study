- plot() - 다양한 기능을 포함하고 있어서, 임의의 개수의 인자를 받을 수 있다.  
x, y 값 인자에 대해 선의 색상과 형태를 지정하는 포맷 문자열을 세번째 인자에 입력할 수 있다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')  
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b-')  
#'ro'는 빨간색 ('red')의 원형 ('o') 마커를 의미
#'b-‘는 파란색 ('blue')의 실선 ('-')을 의미
plt.axis([0, 6, 0, 20])   #[0, 6] = x축 시작값과 마지막값, [-, 20] = y축의 시작값과 마지막 값
plt.show()
```

• Matplotlib에서는 일반적으로 NumPy 어레이를 이용  
모든 시퀀스는 내부적으로 NumPy 어레이로 변환되어 처리됨  
다양한 스타일을 갖는 여러 개의 곡선을 하나의 그래프로 출력

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0., 5., 0.2) # step을 0.2로 지정, 200ms 간격으로 균일한 샘플된 시간

#1. x축 t y축도 t 2.x축은 r--,  y축은 t 3 스타일은 네모. x축은 t**3(3제곱), y축은 g^  스타일은 삼각형
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')  
plt.show()
```

- xlabel(), ylabel() 함수를 사용하면 그래프의 X, Y축에 대한 레이블을 표시할 수 있다.  
labelpad 파라미터는 축 레이블의 여백 (Padding)을 지정  
fontdict 파라미터를 사용하면 축 레이블의 폰트 스타일을 설정

```python
import matplotlib.pyplot as plt

#딕셔너리 font1,2 지정
font1 = {'family': 'serif',
'color': 'b',
'weight': 'bold',
'size': 14 }
font2 = {'family': 'fantasy',
'color': 'deeppink',
'weight': 'normal',
'size': 'xx-large' }


plt.plot([1, 2, 3, 4], [2, 3, 5, 10])

#xlabel=x축의 이름은 'x-axis', labelpad=축으로부터 여백크기 fondict=폰트를 어떻게 출력할지(색상, 사이즈, 웨이트 등) 
plt.xlabel('X-Axis', labelpad=15, fontdict=font1)
#plt.xlabel('X-Axis', labelpad=10)  #이렇게도 확인해보기 (아래에 확인)

plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)
#plt.ylabel('Y-Axis', labelpad=30)
plt.show()
```

- 범례 (Legend)는 그래프에 데이터의 종류를 표시하기 위한 텍스트  
legend() 함수를 사용해서 그래프에 범례를 표시할 수 있다.  
plot() 함수에 label 문자열을 지정하고, matplotlib.pyplot 모듈의 legend() 함수를 호출  
legend() 함수의 loc 파라미터를 이용해서 범례가 표시될 위치를 설정  
loc 파라미터는 문자열로 지정할 수도 있고, 숫자 코드를 사용할 수도 있다.  
loc=(0.0, 0.0)은 데이터 영역의 왼쪽 아래, loc=(1.0, 1.0)은 데이터 영역의 오른쪽 위 위치

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

#legend로 범례 위치 지정 가능
#plt.legend(loc=(0.0, 0.0))  #범례의 위치가 0,0인 것 > 왼쪽 아래(시작점)
#plt.legend(loc=(0.5, 0.5))  #범례의 위치가 0.5,0.5인 것 > 중간
#plt.legend(loc=(1.0, 1.0))  #범례의 위치가 1.0,1.0인 것 > 오른쪽 위
plt.legend(loc='lower right') #범례의 위치가 오른쪽 아래(안에 문구는 바꿔도 됨)
plt.show()
```

- legend() 함수의 ncol 파라미터는 범례에 표시될 텍스트의 열의 개수를 지정  
기본적으로 범례 텍스트는 1개의 열로 표시된다  
legend() 함수의 fontsize 파라미터는 범례에 표시될 폰트의 크기를 지정  
frameon 파라미터는 범례 텍스트 상자의 테두리를 표시할지 여부를 지정  
frameon=False로 지정하면 테두리가 표시되지 않음  
shadow 파라미터를 사용해서 텍스트 상자에 그림자를 표시할 수 있다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# plt.legend(loc='best') # ncol = 1 => 범례 개수1개의 열
#ncol=범례 열 개수2개, fontsize =폰트사이즈, frameon=true면 범례 테두리가 존재, shadow=true면 범례에 그림자가 존재
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)
plt.show()
```

- 선 종류 - 포맷 문자열을 사용  
|  - (Solid), -- (Dashed), : (Dotted), -. (Dash-dot)

```python
import matplotlib.pyplot as plt

#선의 스타일을 지정, label= 범례의 이름
plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')  
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

- linestyle 지정  
plot() 함수의 linestyle 파라미터 값을 직접 지정할 수 있다.  
포맷 문자열과 같이 ‘solid’, ‘dashed’, ‘dotted’, dashdot’ 네가지의 선 종류를 지정할 수 있다.

```python
import matplotlib.pyplot as plt

#linestyle에 선이름을 지정한 것('-'이렇게 표현하는 대신)
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0',
label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0',
label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0',
label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0',
label="'dashdot'")

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])

plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
```

####그래프 영역 채우기
- 그래프의 특정 영역을 색상으로 채워서 강조할 수 있다.

- fill_between() - 두 수평 방향의 곡선 사이를 채움

- fill_betweenx() - 두 수직 방향의 곡선 사이를 채움

- fill() - 다각형 영역을 채움

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]
plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

#네 점 (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역이 채워짐(수평 방향의 곡선사이)
plt.fill_between(x[1:3], y[1:3], alpha=0.5)  #alpha=불투명도

# fill_between() 사용, 네 점 (x[2], y[2]), (x[3], y[3]), (0, y[2]), (0, y[3])을 잇는 영역이 채워짐
#plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.2) #(수직 방향의 곡선사이)
# fill_betweenx() 사용
plt.show()
```

#### 축 스케일 지정하기
- xscale(), yscale() 함수를 사용해서 그래프의 축 스케일을 다양하게 지정
- 축은 기본적으로 ‘linear’ 스케일로 표시, ‘log’, ‘symlog’, ‘logit’으로 변경할 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

#linspace 0부터 5사이의 난수를 생성 exp= 제곱
x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('linear')
#plt.yscale('log')

plt.show()
```

#### 그리드 설정하기

color, alpha, linestyle 파마리터를 사용해서 그리드 선의 스타일을 설정  
which 파라미터를 ‘major’, ‘minor’, ‘both’ 등으로 사용하면 주눈금, 보조눈금에 각각 그리드를 표시

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2) #linewidth =  라인두께
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9) #markersize= 마커사이즈, ^ =마커 삼각형

#grid = 격자표시 True면 기본 격자
plt.grid(True)
#plt.grid(True, axis='y')
#plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')

plt.show()
```

눈금 스타일
- 틱 (Tick)은 그래프의 축에 간격을 구분하기 위해 표시하는 눈금

- xticks(), yticks(), tick_params() 함수를 이용해서 그래프에 눈금을 표시

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 2, 0.2) #0부터 2까지 0.2단위로 

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='forestgreen', marker='^', markersize=9)
#plt.xticks([0, 1, 2])
#plt.yticks(np.arange(1, 6))
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.show()
```

- tick_params() 함수를 사용하면 눈금의 스타일을 다양하게 설정

- axis는 설정이 적용될 축을 지정 {‘x’, ‘y’, ‘both’} 중 선택

- direction을 ‘in’, ‘out’으로 설정하면 눈금이 안/밖으로 표시 {‘in’, ‘out’, ‘inout’} 중 선택

- length는 눈금의 길이를 지정

- pad는 눈금과 레이블과의 거리를 지정

- labelsize는 레이블의 크기를 지정

- labelcolor는 레이블의 색상을 지정

- top/bottom/left/right를 True/False로 지정하면 눈금이 표시될 위치를 선택

- width는 눈금의 너비를 지정

- color는 눈금의 색상을 지정

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

#direction= in or out은 눈금을 안/밖으로, top은 눈금이 위로 color는 눈금색상
plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=14, labelcolor='green',
top=True)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=12, width=2, color='r')

plt.show()
```

####타이틀 설정하기
- title() -그래프의 타이틀 (Title)을 설정

- loc 파라미터를 ‘right’로 설정하면, 타이틀이 그래프의 오른쪽 위에 나타남 {‘left’, ‘center’, ‘right’}

- pad 파라미터는 타이틀과 그래프와의 간격을 포인트 단위로 설정

- fontdict 파라미터에 딕셔너리 형태로 폰트 스타일을 설정

- ‘fontsize’는 포인트 단위의 숫자를 입력하거나 ‘smaller’, ‘x-large’ 등의 상대적인 설정

- ‘fontweight’에는 {‘normal’, ‘bold’, ‘heavy’, ‘light’, ‘ultrabold’, ‘ultralight’}와 같이 설정

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
#'Graph Title' 뒤에 loc,pad,fontdict없이 이름만 적히면 가운데에 나타남 
plt.title('Graph Title', loc='right', pad=20)

title_font = {
'fontsize': 16,
'fontweight': 'bold'
}

plt.title('Graph Title', fontdict=title_font, loc='left', pad=20)
plt.show()
```

- title() - 타이틀을 나타내는 Matplotlib text 객체를 반환

- get_position()과 get_text() - 텍스트 위치와 문자열 반환

```python
import numpy as np

a = np.arange(0, 2, 0.2)
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='forestgreen', marker='^', markersize=9)
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
title_right = plt.title('Graph Title', loc='right', pad=20)
title_font = {
'fontsize': 16,
'fontweight': 'bold'
}

title_left = plt.title('Graph Title', fontdict=title_font, loc='left', pad=20)
```

```python
print(title_left.get_position())  #타이틀의 텍스트 위치를 출력
print(title_left.get_text())  #타이틀의 텍스트 내용을 반환
print(title_right.get_position())  #오른쪽 위에있는 타이틀이니 (1.0, 1.0)
print(title_right.get_text())

plt.show()
```

####수직선/수평선 표시
- axhline(): 축을 따라 수평선을 표시, 첫번째 인자는 x 값으로서 수직선의 위치  
두, 세번째 인자는 ymin, ymax 값으로서 0에서 1 사이의 값을 입력  
0은 아래쪽 끝 (X축), 1은 위쪽 끝을 의미

- axvline(): 축을 따라 수직선을 표시 , 첫번째 인자는 x 값으로서 수평선의 위치,  
두, 세번째 인자는 xmin, xmax 값으로서 0에서 1 사이의 값을 입력  
0은 왼쪽 끝 (Y 축), 1은 오른쪽 끝을 의미

- hlines(): 지정한 점을 따라 수평선을 표시, y, xmin, xmax를 순서대로 입력하면, 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시

- vlines(): 지정한 점을 따라 수직선을 표시, x, ymin, ymax를 순서대로 입력하면, 점 (x, ymin)에서 점 (x, ymax)를 따
라 수평선을 표시

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='forestgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

#axh =수평선을 그음, ahv=수직선을 그음
plt.axhline(1, 0, 0.55, color='gray', linestyle='--', linewidth='1')
plt.axvline(1, 0, 0.16, color='lightgray', linestyle=':', linewidth='2')

plt.axhline(5.83, 0, 0.95, color='gray', linestyle='--', linewidth='1')
plt.axvline(1.8, 0, 0.95, color='lightgray', linestyle=':', linewidth='2')

plt.show()
```

- plt.figure()는 새로운 figure를 생성 => figure객체 생성  

- add.subplot()는 그려지는 그래프개수만클 subplot을 만듬  
add_subplot(2, 1, 1) 2행1열에서 1번 인덱스위치에 그래프를 그리라는 뜻  
ax1 = add_subplot(2, 1, 1)

이런식으로 



```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

n = 100
#x축은 0,20까지 y축은 0,20까지 z축은 0,50까지
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50   
cmin, cmax = 0, 2

xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')
plt.show()
```

####히스토그램 (Histogram)
- 도수분포표를 그래프로 나타낸 것의로 가로축은 계급, 세로축은 도수 (횟수나 개수 등)

- hist() - bins 파라미터는 히스토그램의 가로축 구간의 개수를 지정 , 기본값 10

- 누적 히스토그램 - cumulative파라미터를 True로 지정

- histtype은 히스토그램의 종류를 지정

```python
import matplotlib.pyplot as plt

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

#plt.hist(weight, label='bins=10')
#plt.hist(weight, bins=30, label='bins=30') #bins=30 가로축 구간 갯수는 30
plt.hist(weight, cumulative=True, label='cumulative=True') #cumulative= 누적히스토그램
plt.hist(weight, cumulative=False, label='cumulative=False')  #false라서 누적하지 않은 히스토그램
plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
weight2 = [52, 67, 84, 66, 58, 78, 71, 57, 76, 62, 51, 79, 69, 64, 76, 57, 63, 53, 79, 64, 50, 61]

plt.hist((weight, weight2), histtype='bar')
plt.title('histtype - bar')  #histtype -히스토그램 종류
plt.figure()

plt.hist((weight, weight2), histtype='barstacked')  #barstacked 히스토그램이 겹쳐진 것
plt.title('histtype - barstacked')
plt.figure()

plt.hist((weight, weight2), histtype='step') #step 히스토그램이 계단형
plt.title('histtype - step')
plt.figure()

plt.hist((weight, weight2), histtype='stepfilled')  #stepfilled 히스토그램이 계단형인데 색이 채워짐
plt.title('histtype - stepfilled')
plt.show()
```

####에러바 표시
- 에러바 (Errorbar, 오차막대)는 데이터의 편차를 표시하기 위한 그래프 형태

- errorbar() - 데이터의 편차를 나타내는 리스트 yerr을 입력

- 데이터 포인트를 기준으로 비대칭인 편차를 표시하려면, (2, N) 형태의 값들을 입력 (N: 데이터 개수)  
첫번째 튜플의 값들은 아래 방향 편차, 두번째 튜플의 값들은 위 방향 편차를 나타냄

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
#yerr 데이터 편차를 지정 첫 튜플은 아래방향편차, 두번째 튜플은 윗방향편차
yerr = [(2.3, 3.1, 1.7, 2.5), (1.1, 2.5, 0.9, 3.9)]

plt.errorbar(x, y, yerr=yerr)
plt.show()
```

- uplims, lolims를 사용해서 상한/하한 기호를 선택적으로 표시

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 5)
y = x**2
yerr = np.linspace(0.1, 0.4, 4)

plt.errorbar(x, y + 4, yerr=yerr)
plt.errorbar(x, y + 2, yerr=yerr, uplims=True, lolims=True)

upperlimits = [True, False, True, False]
lowerlimits = [False, False, True, True]
plt.errorbar(x, y, yerr=yerr, uplims=upperlimits, lolims=lowerlimits)
plt.show()
```

####이중 Y축 표시

- set_xlabel(), set_ylabel() 메서드는 각 축에 대한 레이블을 표시하도록 합니다.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12
x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1
fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
ax1.plot(x, y1, color='green')

ax2 = ax1.twinx() #x축을 공유하는 새로운 Axes 객체를 만듭니다.
ax2.set_ylabel('2nd Y-Axis')
ax2.plot(x, y2, color='deeppink')
plt.show()
```

- set_zorder() 메서드를 사용해서 그래프가 표시될 순서를 지정, z-축 방향의 순서를 지정

```python
x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])
fig, ax1 = plt.subplots()

ax1.plot(x, y1, '-s', color='green', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax1.set_ylim(0, 18)
ax1.set_xlabel('Year')
ax1.set_ylabel('Price ($)')
ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx()
ax2.bar(x, y2, color='deeppink', label='Demand', alpha=0.7, width=0.7)
ax2.set_ylim(0, 18)
ax2.set_ylabel(r'Demand ($\times10^6$)')
ax2.tick_params(axis='y', direction='in')

ax1.set_zorder(ax2.get_zorder() + 10) # zorder가 낮을수록 먼저 그려지고
ax1.patch.set_visible(False) # zorder가 높을수록 나중에 그려집니다.
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
```

####박스 플롯
- 박스-위스커 플롯 (Box-Whisker plot)은 수치 데이터를 표현하는 하나의 방식
- 다섯 가지 요약 수치를 사용해서 그려집니다. - 최소값, 제 1사분위 수 (Q1), 제 2사분위 수 또는 중위수 (Q2), 제 3사분위 수 (Q3), 최대값
- 가운데 박스 모양으로부터 그려지는 중심선 - 수염 (Whisker)
- boxplot() 함수는 기본적으로 박스의 위쪽 경계로부터 Q3 + whis * (Q3-Q1)까지, 박스의 아래쪽 경계로부터 Q1 - whis * (Q3-Q1)까지 수염 (Whisker)을 나타냅니다.

```python
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c])
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
```

####애니메이션 사용
- matplotlib.animation 모듈의 FuncAnimation 클래스를 사용해서 그래프를 애니메이션으로 표현

- ḏ ۃ Ӓ 인자로 Figure 객체를, 두번째 인자로는 프레임마다 반복해서 호출할 함수를 지정

- frames 키워드 인자에는 반복 가능한 객체를 입력

- np.linspace(a, b, n)는 a부터 b 사이를 균일한 간격으로 나누는 숫자 n개를 반환

- update() 함수에 frames에 제공한 값들이 순서대로 제공되어서 프레임을 구성

```python
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.2, 1.2)
x, y = [], []
line, = plt.plot([], [], 'bo')

def update(frame):
  x.append(frame)
  y.append(np.sin(frame))
  line.set_data(x, y)
  return line,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128))
```
